```python
'''
func: convert video naming to a specific naming convention: video_synthesisType_compressionType.mp4 --> video,path_name.parts[-3],path_name.parts[-2]
input:
    i.curr_path: string: the folder path of the video (Eg: ./Dataset/manipulated_sequences/FaceSwap/c40/videos)
    ii.target: string: the final directory to move the video file to
'''
def video_naming(curr_path,target):
    #convert to Path
    path = Path(curr_path)
    #extract relevant information from file path
    synthType = path.parts[-4]
    compType = path.parts[-3]
    video = path.parts[-1]
    #split .mp4 to get video name
    name = video.split('.')[0]
    
    #target path
    tar_path = os.path.join(target+'videos/')
    newName = os.path.join(tar_path+name+'_'+synthType+'_'+compType+'.mp4')
    try:
        #renaming file and moving it to the target folder
        os.rename(curr_path,newName)
    except Exception as e:
        print(e)
```

```python
'''
func: convert video naming to a specific naming convention: video_synthesisType_compressionType.mp4 --> video,path_name.parts[-3],path_name.parts[-2]
input: none
'''
def move_files():
    print('Move')
    #move and rename files to create the required structure
    for target,source in ref_dir.items():
        #loop through the different synthesis folders
        for val in source:
            path = os.path.join(val+'/**/**/*.mp4')
            files = glob.glob(path, recursive = True) 
            #for each file call video_naming function and call rename and move the file
            for file in files:
                video_naming(file, target)
            print('Files renamed and moved for',val,'to',target)
```

## Pre-Processing Phase

Once the folder structure is ready, the videos are pre-processed:
1. Extracting Frames
2. Adding Noise to the images
3. Face Cropping 

*Note that image re-sizing is not done in this phase. This will vary basis the transfer learning models used for feature extraction*

```python
'''
func: while extracting frames, make directory for each video to store the respective frames
input:
    i.video: string: the name of the video without the file extension
    ii.target: string: the file path for the frames folder (Eg: ./Dataset/FaceSwap/frames)
'''
def make_dir(video,target):
    dirName = target+'/'+video ##target directory name
    try:
        # Create target Directory
        os.mkdir(dirName)
    except FileExistsError:
        print("Directory " , dirName ,  " already exists")
```

```python
'''
func: captures frames from the video, adds noise to the video and saves in the given folder
input:
    i.vid_path: string: the name of the video 
    ii.target: string: the file path for the frames folder (Eg: ./Dataset/FaceSwap/frames)
    iii. req_fps: int: required video fps for frame extraction. No.of frames extracted will be based on this
    iv. faceCrop: bool: value to determine if face cropping is necessary or no
'''
def FrameCapture(vid_path,target,req_fps,faceCrop): 
    # Path to video file 
    vidObj = cv2.VideoCapture(vid_path)
    
    #Reduce the FPS
    fps = vidObj.get(cv2.CAP_PROP_FPS)
    frame_count = vidObj.get(cv2.CAP_PROP_FRAME_COUNT)
    if ((frame_count!=0) & (fps !=0)):
        req_frame_count = (req_fps*frame_count)/fps
        count_check = frame_count/req_frame_count
    # Used as counter variable 
    count = 0
    # checks whether frames were extracted 
    success = 1
    
    #Make frame folder
    path = Path(vid_path)
    video = path.parts[-1]
    video = video.split('.')[0]
    make_dir(video,target)

    while success: 
        # vidObj object calls read 
        success, image = vidObj.read() 
        if np.shape(image) == ():
            continue
            #print('Empty Image Error')
        elif ((count%count_check) == 0):
            framepath = os.path.join(target+'/'+video+"/"+video+"_frame%d.jpg" % count)
            img = AddNoise(image)
            if faceCrop == True:
                crop_img = CropFace(img)
                try:
                    cv2.imwrite(framepath, crop_img)
                except Exception as e:
                    print(e)
            else:
                cv2.imwrite(framepath, img)

        count += 1
```

```python
'''
func: add gaussian noise to the frames 
input:
    i.image: object: image object video frame
'''
def AddNoise(image):
    # Add gaussian noise to the image.
    noise_img = random_noise(image, mode='gaussian')

    # The above function returns a floating-point image
    # on the range [0, 1], thus we changed it to 'uint8'
    # and from [0,255]
    noise_img = np.array(255*noise_img, dtype = 'uint8')
    return noise_img
    #cv2.imwrite(image_path, noise_img)
```

```python
'''
func: locate face using CV2 and crop it. 
input:
    i.image: object: image object video frame
'''
def CropFace(image):
    
    #reading the image
    #image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    #finding a face
    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=3,
        minSize=(30, 30)
    )
            
    #cropping the image
    for (x, y, w, h) in faces:
        #cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        roi_color = image[y-20:y + h + 20, x-20:x + w + 20] #the + and - 20 gives a little bit of border around the face
        return roi_color
```

```python
'''
func: main pre-processing function 
input:
    i.vid_dir: string: video filepath
    ii. extension: string: 2 possible values frames or face_crop
    iii. faceCrop: bool: boolean value to determine if face cropping is needed
'''
#Main Pre-Process Function
def Pre_Process(vid_dir, extension, faceCrop):
    for video in vid_dir:
        path = os.path.join(video+'videos/*.mp4')
        files = glob.glob(path, recursive = True) 
        target = os.path.join(video+extension)
        for file in files:
            FrameCapture(file,target,0.2,faceCrop)
        print('Pre-Process Complete')
```

```python
#performing pre-processing with face crop
for technique in final_dir:
    print(technique)
    Pre_Process([technique,''],'face_crop',True)
```
